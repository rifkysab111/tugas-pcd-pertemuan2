```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('sahilif24b.jpeg', 0)

def quantize(image, bits):
    levels = 2**bits
    interval = 256 // levels

    quantized = (image // interval) * interval
    return quantized

bits_2 = quantize(img, 2)
bits_4 = quantize(img, 4)
bits_8 = quantize(img, 8)

titles = ['Original (8-bit)', 'Kuantisasi 4-bit (16 Level)', 'Kuantisasi 2-bit (4 Level)']
images = [bits_8, bits_4, bits_2]

plt.figure(figsize=(15, 5))
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
