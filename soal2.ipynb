```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('kahim24.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

resolutions = [512, 256, 128, 64]
plt.figure(figsize=(15, 5))

for i, res in enumerate(resolutions):
    resized_img = cv2.resize(img_rgb, (res, res), interpolation=cv2.INTER_LINEAR)

    plt.subplot(1, 4, i+1)
    plt.imshow(resized_img)
    plt.title(f"Resolusi {res}x{res}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```
